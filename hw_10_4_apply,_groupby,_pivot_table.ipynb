3. аналіз покупок по дням тижня

```python
# перш ніж працювати з запитом перевіримо тип потрбної колонки
df['Date'].dtype
```

```python
# міняємо рядковий тип на потрібний для подальших перетворень
df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%Y")
```

```python
# перевіряємо чи змінився тип
df['Date'].dtype
```

```python
# створюємо колонку з назвою дня тижня
df['Day'] = df['Date'].dt.day_name()
```

```python
df [['Date', 'Day']]. head()
```

```python
# створюємо півот тейбл для подальшого аналізу по днях
summary = pd.pivot_table(
    df,
    index='Day',
    values=['Invoice ID', 'Total'],
    aggfunc={
        "Invoice ID": 'count',
        "Total": ['sum', 'mean']
    }
)
summary
```

```python
summary.columns = ["Orders", "Revenue", "Avg_Check"]
summary
```

```python
# змінюємо порядок днів тижня
order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
summary = summary.reindex(order)

summary
```

Згідно з данимив таблиці бачимо, що найактивніший день покупок це субота, найменше покупок здійсеюють в понеділок. При цьому в суботу є і навищий середній чек. Це повязано з тим, що в вихідний день лбюди більше приділяють увагу закупам на тиждень. Тобто на суботу потрібно щоб були відповідні запаси товарів, можна більше виставляти товарів нових або тих, що гірше продаються. Відповідно в понеділок, коли продажі менші і по кількості і по сумах чеків можна пропонувати клієнтам вигідні пропозиція дня. і т.д.
